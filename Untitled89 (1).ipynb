```python
import numpy as np


x = [1, 2, 3]


def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)


Wf, Whf, bf = 0.5, 0.1, 0
Wi, Whi, bi = 0.6, 0.2, 0
Wc, Whc, bc = 0.7, 0.3, 0
Wo, Who, bo = 0.8, 0.4, 0


h = 0
c = 0

for t in range(len(x)):
    
    xt = x[t]

    f = sigmoid(Wf*xt + Whf*h + bf)
    i = sigmoid(Wi*xt + Whi*h + bi)
    c_tilde = tanh(Wc*xt + Whc*h + bc)

    c = f*c + i*c_tilde

    o = sigmoid(Wo*xt + Who*h + bo)
    h = o * tanh(c)

    print("Time step", t+1)
    print("Forget gate =", f)
    print("Input gate =", i)
    print("Candidate state =", c_tilde)
    print("Cell state =", c)
    print("Output gate =", o)
    print("Hidden state =", h)
    print()


Wy = 1.0
by = 0.0

y_pred = Wy*h + by

print("Predicted next value =", y_pred)
```
